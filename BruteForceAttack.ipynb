```python
import itertools
```

```python
def check_password(password,actual_password):
    return password == actual_password
```

```python
def brute_force_attack(actual_password):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    password_length = 6  # Maximum password length

    for length in range(1, password_length + 1):
        # Generate all possible combinations of characters of length 'length'
        combinations = itertools.product(alphabet, repeat=length)

        for combination in combinations:
            # Join the characters to form a password
            password = ''.join(combination)
            print("Trying password:", password)

            # Check if the generated password matches the actual password
            if check_password(password,actual_password):
                print("Password found:", password)
                return

    print("Password not found.")
```

```python
actual_password=input()
brute_force_attack(actual_password)
```
